
## Wrangle

- Data wrangling, or data munging, is the process of cleaning, structuring, and enriching raw data into a usable format for analysis. By transforming messy datasets into high-quality assets, it ensures that organizations can make accurate, data-driven decisions and improve the performance of machine learning models.
- The process is essential for handling the "six Vs" of big data—volume, velocity, variety, veracity, value, and variability. Effective wrangling reduces the time spent on data preparation, allowing data scientists to focus more on uncovering insights and less on fixing formatting errors.

Common Data Wrangling Steps:

- **Discovery:** Identifying and understanding what the data contains and how it can be used to meet specific business objectives.
- **Structuring:** Organizing raw data into a consistent format, such as aligning columns and rows, to make it compatible with analytical tools.
- **Cleaning:** Removing outliers, fixing null values, and resolving inconsistencies to ensure the dataset is accurate and reliable.
- **Enriching:** Augmenting the existing dataset by adding extra variables or integrating data from other sources to provide more context.
- **Validating:** Verifying that the data is consistent, high-quality, and follows security protocols before it is used for modeling.
- **Publishing:** Delivering the final, processed dataset into a database or warehouse where it can be accessed for analysis and reporting.


Data Science: is a branch of computer science where we study how to store, use and analyze data for deriving information from it.
