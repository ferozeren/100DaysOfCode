
## Computations

- In the context of data science, computation is the application of mathematical models, advanced algorithms, and numerical analysis to process, analyze, and extract actionable insights from complex datasets.

- While native Python is highly readable and flexible, it is an *interpreted* language. This makes it inefficient for the high-speed, large-scale numerical processing required for massive datasets out of the box.

- To solve this, the Python data stack relies on specialized computational libraries. These tools wrap highly optimized, low-level compiled languages (like C or Fortran) in simple Python syntax, giving data scientists the ease of Python with the processing speed of C.


### Numpy

- Fast and versatile, the NumPy vectorization, indexing, and broadcasting concepts are the de-facto standards of array computing today.

- Offers comprehensive mathematical functions, random number generators, linear algebra routines, Fourier transforms, and more.

- Supports a wide range of hardware and computing platforms, and plays well with distributed, GPU, and sparse array libraries.

- Performant (optimized C code as it's core) and easy to use (High level syntax).
