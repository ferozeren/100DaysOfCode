
## Blueprint

- Data visualization is the representation of data through use of common graphics, such as charts, plots, infographics and even animations. These visual displays of information communicate complex data relationships and data-driven insights in a way that is easy to understand.

- Data visualization is classified into four functional areas: Idea Generation (brainstorming), Idea Illustration (explaining processes), Visual Discovery (finding trends), and Everyday Dataviz (storytelling/reporting).

Types of data Visualizaions:

- **Tables:** Use rows and columns to show detailed, structured data comparisons.
- **Pie & Stacked Bar Charts:** Compare "parts-of-a-whole" to show relative sizes of components.
- **Line & Area Charts:** Track changes and trends over time, often for predictive analytics.
- **Histograms:** Display data distribution and frequency to easily identify outliers.
- **Scatter & Bubble Plots:** Reveal relationships between variables and support regression analysis.
- **Heat & Tree Maps:** Visualize behavioral data by location or compare hierarchical proportions via shape size.

### Matplotlib

- Comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible.

- Provide ways to:
    - Create publication quality plots.
    - Make interactive figures that can zoom, pan, update.
    - Customize visual style and layout.
    - Export to many file formats.
    - Embed in JupyterLab and Graphical User Interfaces.
    - Use a rich array of third-party packages built on Matplotlib.
